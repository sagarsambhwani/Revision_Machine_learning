# Unsupervised Learning

Unsupervised learning is a type of machine learning where the model is trained on data that has no labeled outputs or predefined target variables. The goal is to discover hidden patterns, groupings, or relationships in the data without prior knowledge of the categories or labels. Unlike supervised learning, which works with input-output pairs (e.g., classification and regression tasks), unsupervised learning works with datasets where only input features are available.

## 1. K-means Clustering

=> Partitions the data into K clusters based on the nearest mean (centroid).

=> Use case: Customer segmentation, document clustering.

```python
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load Wine dataset
wine = load_wine()
X = wine.data

# Apply K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.title('K-means Clustering on Wine Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()
```

## 2. Hierarchical Clustering:

=> Creates a hierarchy of clusters (dendrogram).

=> Agglomerative: Starts with individual points and merges.

=> Divisive: Starts with one large cluster and splits.

=> Use case: Gene analysis, grouping similar products.

```python
from sklearn.datasets import load_wine
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Load Wine dataset
wine = load_wine()
X = wine.data

# Perform Agglomerative Clustering (Hierarchical Clustering)
hierarchical = AgglomerativeClustering(n_clusters=3)
labels = hierarchical.fit_predict(X)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolor='k')
plt.title('Hierarchical Clustering on Wine Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster')
plt.show()

# Create a Dendrogram
Z = linkage(X, 'ward')  # Using Ward's method for linkage
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Dendrogram for Wine Dataset')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

# ==> Dimensionality Reduction Techniques:

## 1. PCA (Principal Component Analysis):

=> Reduces dimensions by finding new axes with maximum variance.

=> Use case: Visualizing high-dimensional data, speeding up models.

```python
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load Wine dataset
wine = load_wine()
X = wine.data

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the PCA-transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=wine.target, cmap='plasma', edgecolor='k')
plt.title('PCA on Wine Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Wine Class')
plt.show()
```

## 3.LDA (Linear Discriminant Analysis):

=> Maximizes class separation in labeled datasets.

=> Use case: Feature extraction for classification problems.


```python
from sklearn.datasets import load_wine
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt

# Load Wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Apply LDA to reduce to 2 components
lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)

# Plot the LDA-transformed data
plt.figure(figsize=(8, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='plasma', edgecolor='k')
plt.title('LDA on Wine Dataset')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')
plt.colorbar(label='Wine Class')
plt.show()
```
